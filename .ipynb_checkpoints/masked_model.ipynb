# First Attempt CNN Masked Model
- Multi-class model and binary classification 
- Learning Rates to prevent overfitting 

```python
import tensorflow as tf
from tensorflow.keras import layers, regularizers

def make_model(input_shape, num_classes):
    model = tf.keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Rescaling(1.0 / 255),
        layers.Conv2D(64, 3, strides=2, padding="same", activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(128, 3, strides=2, padding="same", activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(2, activation='sigmoid')
    ])
    if num_classes == 2:
        model.add(layers.Dense(1, activation='sigmoid'))
    else:
        model.add(layers.Dense(num_classes, activation=None))

    return model

image_size = (224, 224)  # Update with your image size
model = make_model(input_shape=image_size + (3,), num_classes=2)  # Update num_classes if necessary

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),  # Adjust learning rate
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=40, batch_size=12, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
def plotTrainingHistory():
    # Training History
    plt.figure(figsize=(8, 8))
    epochs_range = range(1, 41)  
    plt.plot(epochs_range, history.history['accuracy'], label="Training Accuracy")
    plt.plot(epochs_range, history.history['val_accuracy'], label="Validation Accuracy")
    plt.axis(ymin=0.4, ymax=1)
    plt.grid()
    plt.title('Model Accuracy: CNN-KNN')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend(['train', 'validation'])
    plt.show()
plotTrainingHistory()
```

# Second Attempt Masked Model CNN
- Model Complexity Increases, Increase in Convolutional Layers
- L2 Regularisers

```python
import tensorflow as tf
from tensorflow.keras import layers, regularizers

def make_model(input_shape, num_classes):
    model = tf.keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Rescaling(1.0 / 255),
        layers.Conv2D(64, 3, strides=2, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(128, 3, strides=2, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(256, 3, strides=2, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(512, 3, strides=2, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(728, 3, strides=2, padding="same", activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
        layers.Dropout(0.5),
        layers.Dense(2, activation='sigmoid')
    ])
    if num_classes == 2:
        model.add(layers.Dense(1, activation='sigmoid'))
    else:
        model.add(layers.Dense(num_classes, activation=None))

    return model

image_size = (224, 224)  # Update with your image size
model = make_model(input_shape=image_size + (3,), num_classes=2)  # Update num_classes if necessary

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),  # Adjust learning rate
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

model.summary()
```

# Final Attempt: Masked Model
- L2 Regularisers
- Increased Complexities
- ReSscalling
- Dropouts

```python
import tensorflow as tf
from tensorflow.keras import layers, regularizers

def make_model(input_shape, num_classes):
    model = tf.keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Rescaling(1.0 / 255),
        layers.Conv2D(64, 3, strides=2, padding="same", activation='relu',kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(128, 3, strides=2, padding="same", activation='relu',kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(256, 3, strides=2, padding="same", activation='relu',kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(512, 3, strides=2, padding="same", activation='relu',kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(728, 3, strides=2, padding="same", activation='relu',kernel_regularizer=regularizers.l2(0.01)),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(2, activation='sigmoid')
    ])
    if num_classes == 2:
        model.add(layers.Dense(1, activation='sigmoid'))
    else:
        model.add(layers.Dense(num_classes, activation=None))

    return model

image_size = (224, 224)  # Update with your image size
model = make_model(input_shape=image_size + (3,), num_classes=2)  # Update num_classes if necessary

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),  # Adjust learning rate
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

model.summary()
```

- This CNN model is created:
  - Convolutional layers followed by batch normalisation, relu activation, and max pooling layers
  - Global average pooling layer reduces spatial dimensions
  - Dropout layers is added for regularisation of the model
  - Sigmoid activation is for binary classification but can be change for multi-class classification provided requirements of the tasks

```python
history = model.fit(X_train, y_train, epochs=40, batch_size=12, validation_split=0.2)
```

- In this specific output, it seems that the model starts with relatively high training and validation accuracy.
- However, as the training progresses, the validation accuracy remains stagnant at around 52.78%, while the training accuracy increases to around 97.16%.
- The increasing gap between training and validation accuracy suggests that the model is overfitting to the training data, meaning it's memorizing the training examples rather than learning general patterns that can be applied to unseen data.
- This could be due to the sample data being too small and unbalanced due to poor generalisation performance
- Learning rate of the model could be too high, causing overfitting
- Model could be too complex

- After fine-tuning the model it was found that the learning-rate of the model was too fast. Therefore, reducing the learning rate allowed for the model to learn the specific patterns of the image set instead of recalling previous patterns learnt.
- Additional dropout layers were added for regularisation.
- To improve the performance from validation accuracy of around 76%, hyperparameter tuning should be investigated.

```python
import time
from sklearn.metrics import confusion_matrix
import seaborn as sns
from scipy.stats import pearsonr

# CNN Alone
def printConfusionMatrix(model, X_test, y_test):
    
    start_time = time.time()
    # Get predictions
    y_pred = model.predict(X_test)
    end_time = time.time()
    print('Inference Time: ', end_time - start_time)
    y_pred_classes = (y_pred > 0.5).astype(int)

    cm = confusion_matrix(y_test, y_pred_classes)
     
    # Calculate Pearson correlation coefficient and p-value
    correlation_coefficient, p_value = pearsonr(y_pred.flatten(), y_pred_classes.flatten())

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title(f'Confusion Matrix: CNN, R-Value: {p_value}')
    plt.show()
    
printConfusionMatrix(model, X_test, y_test)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import warnings
import tracemalloc

warnings.filterwarnings("ignore", category=FutureWarning)

tracemalloc.start()
start_time = time.time()
# Train KNN from Trained Model
with tf.device('/CPU:0'):
    feature_extractor = tf.keras.Model(inputs=model.input, outputs=model.layers[-3].output)
    X_train_features = feature_extractor.predict(X_train)
    X_test_features = feature_extractor.predict(X_test)
    
# Reshape features for KNN
X_train_features_flattened = X_train_features.reshape(X_train_features.shape[0], -1)
X_test_features_flattened = X_test_features.reshape(X_test_features.shape[0], -1)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_features_flattened)
X_test_scaled = scaler.transform(X_test_features_flattened)

# Training KNN
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train_scaled, y_train)

# Predict using KNN
cnn_rnn_y_pred = knn.predict(X_test_scaled)
end_time = time.time()

current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print(f'Current and Peak Memory Consumption {current, peak}')

# Evaluate KNN performance
accuracy = accuracy_score(y_test, cnn_rnn_y_pred)
print("KNN Accuracy:", accuracy)

print('Inference Time: ', end_time - start_time)
```

```python
from sklearn.metrics import classification_report

# CNN-KNN statistic
from sklearn.metrics import matthews_corrcoef

# Set the threshold for binary classification
threshold = 0.5

# Convert probabilities to binary predictions
y_pred_classes = (cnn_rnn_y_pred > threshold).astype(int)

# Generate classification report with handling for zero division
class_report = classification_report(y_test, y_pred_classes, zero_division=0)

print("Classification Report:\n", class_report)

# Assuming y_true contains true labels and y_pred contains predicted probabilities
mcc = matthews_corrcoef(y_test, y_pred_classes > 0.5)  
print("Matthews Correlation Coefficient (MCC):", mcc)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr


# Threshold probabilities to obtain class labels
y_pred_binary = (cnn_rnn_y_pred >= 0.5).astype(int)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred_binary)

# Calculate Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(cnn_rnn_y_pred.flatten(), y_pred_binary.flatten())

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title(f'Confusion Matrix: CNN-KNN, R-Value: {p_value}')
plt.show()

```

```python
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# Define a range of values for n_neighbors
neighbors_range = range(1, 11)  # Example range, adjust as needed

# Initialize lists to store accuracy values
accuracies = []

# Iterate over different values of n_neighbors
for n in neighbors_range:
    # Initialize and fit KNN classifier
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train_scaled, y_train)

    # Predict using KNN
    y_pred = knn.predict(X_test_scaled)

    # Evaluate KNN performance
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot the accuracy vs. n_neighbors
plt.plot(neighbors_range, accuracies, marker='o')
plt.title('KNN Accuracy vs. Number of Neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(neighbors_range)
plt.grid(True)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt


def printTestCalculations(X_test, y_test, model):
    num_samples = 5

    # Randomly select indices of samples from X_test
    sample_indices = random.sample(range(len(X_test)), num_samples)
    
    fig, axes = plt.subplots(1, num_samples, figsize=(20, 4))  

    for i, index in enumerate(sample_indices):
        
        image = X_test[index]
        true_label = y_test[index]

        # predicted_probability being selected
        predicted_probability = model.predict(np.expand_dims(image, axis=0))[0]  

        predicted_probability_scalar = predicted_probability.item() if hasattr(predicted_probability, 'item') else predicted_probability
        predicted_label = 1 if predicted_probability_scalar > 0.5 else 0

        # display images
        axes[i].imshow(image)
        axes[i].axis('off')
        axes[i].set_title(f"True Label: {true_label}\nPredicted Label: {predicted_label} (Probability: {predicted_probability_scalar:.4f})", fontsize=10)

    plt.subplots_adjust(wspace=0.3)  
    plt.tight_layout()
    plt.show()
    
printTestCalculations(X_test=X_test, y_test=y_test, model=model)
```

```python
from sklearn.metrics import precision_recall_curve

def plotPredictionRecall():
    # Compute precision-recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred)

    # Plot precision-recall curve
    plt.figure()
    plt.step(recall, precision, color='b', alpha=0.2, where='post')
    plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title('Precision-Recall Curve: CNN-KNN')
    plt.show()
plotPredictionRecall()
```

```python

def compute_complexity(model):
    complexity = 0
    for layer in model.layers:
        if isinstance(layer, tf.keras.layers.Dense):
            input_size = layer.input_shape[1]  # Number of input features
            output_size = layer.units  # Number of output units
            complexity += input_size * output_size
    return complexity

compute_complexity(model=model) # computation operations to process a single input 
```
