```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import regularizers


image_size = (224, 224)
channels = 3
image_shape = (image_size[0], image_size[1], channels)

base_model = MobileNetV2(
    include_top=False,
    weights="imagenet",
    input_shape=image_shape
)

# Mark the base model as non-trainable
base_model.trainable = False

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),               
    Dense(64, activation='relu'),                
    Dense(32, activation='relu'),               
    Dense(16, activation='relu'),   
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
import matplotlib.pyplot as plt
def plotTrainingHistory():
    # Plot training history
    plt.figure(figsize=(8, 8))
    epochs_range = range(1, 21)  # Assuming you trained for 10 epochs
    plt.plot(epochs_range, history.history['accuracy'], label="Training Accuracy")
    plt.plot(epochs_range, history.history['val_accuracy'], label="Validation Accuracy")
    plt.axis(ymin=0.4, ymax=1)
    plt.grid()
    plt.title('Model Accuracy: MobileNetv2')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend(['train', 'validation'])
    plt.show()
plotTrainingHistory()
```

# Final Attempt: MobileNetv2 
- Additional of L2 Regularisers

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import regularizers


image_size = (224, 224)
channels = 3
image_shape = (image_size[0], image_size[1], channels)

base_model = MobileNetV2(
    include_top=False,
    weights="imagenet",
    input_shape=image_shape
)

# Mark the base model as non-trainable
base_model.trainable = False

# L2 Regularisation is applied to penalize large weight values to prevent overfitting
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(rate=0.5),
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)),  
    Dropout(rate=0.5),              
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),   
    Dropout(rate=0.5),               
    Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)),   
    Dropout(rate=0.5),              
    Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.01)),   
    Dropout(rate=0.5),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
# Train the model using only the training data and specify the validation split
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from keras.optimizers import Adamax

early_stopping = EarlyStopping(
    monitor='val_loss',   # Monitor validation loss
    patience=3,           # Number of epochs with no improvement after which training will be stopped
    restore_best_weights=True  # Restore model weights to the point with the best validation loss
)

history = model.fit(X_train, y_train, batch_size=20, epochs=20, validation_split=0.2)
```

```python
import random
import matplotlib.pyplot as plt


def printTestCalculations(X_test, y_test, model):
    num_samples = 5

    # Randomly select indices of samples from X_test
    sample_indices = random.sample(range(len(X_test)), num_samples)
    
    fig, axes = plt.subplots(1, num_samples, figsize=(20, 4))  

    for i, index in enumerate(sample_indices):
        
        image = X_test[index]
        true_label = y_test[index]

        # predicted_probability being selected
        predicted_probability = model.predict(np.expand_dims(image, axis=0))[0]  

        predicted_probability_scalar = predicted_probability.item() if hasattr(predicted_probability, 'item') else predicted_probability
        predicted_label = 1 if predicted_probability_scalar > 0.5 else 0

        # display images
        axes[i].imshow(image)
        axes[i].axis('off')
        axes[i].set_title(f"True Label: {true_label}\nPredicted Label: {predicted_label} (Probability: {predicted_probability_scalar:.4f})", fontsize=10)

    plt.subplots_adjust(wspace=0.3)  
    plt.tight_layout()
    plt.show()
    
printTestCalculations(X_test=X_test, y_test=y_test, model=model)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def printConfusionMatrix(model, X_test, y_test):
    # Get predictions
    y_pred = model.predict(X_test)
    y_pred_classes = (y_pred > 0.5).astype(int)

    cm = confusion_matrix(y_test, y_pred_classes)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix: MobileNetv2')
    plt.show()
printConfusionMatrix(model, X_test, y_test)
```

```python
from sklearn.metrics import classification_report, matthews_corrcoef

# Set the threshold for binary classification
threshold = 0.5

# Convert probabilities to binary predictions
y_pred_classes = (y_pred > threshold).astype(int)

# Generate classification report with handling for zero division
class_report = classification_report(y_test, y_pred_classes, zero_division=0)

mcc = matthews_corrcoef(y_test, y_pred_classes > 0.5)  
print("Matthews Correlation Coefficient (MCC):", mcc)

print("Classification Report:\n", class_report)
```

```python

def compute_complexity(model):
    complexity = 0
    for layer in model.layers:
        if isinstance(layer, tf.keras.layers.Dense):
            input_size = layer.input_shape[1]  # Number of input features
            output_size = layer.units  # Number of output units
            complexity += input_size * output_size
    return complexity

compute_complexity(model=model) # computation operations to process a single input 
```

```python
def calculate_model_size(model):
    num_params = sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    model_size_bytes = num_params * 4  # Assuming parameters are stored as float32 (4 bytes)
    return model_size_bytes / (1024**2)  # Convert bytes to megabytes

calculate_model_size(model=model)
```
