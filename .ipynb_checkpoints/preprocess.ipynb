```python
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Check Mafa Masked Seperation 
def checkMaskedValidity(masked_folder):
    # jpg and png's
    image_files = [file for file in os.listdir(masked_folder) if file.endswith('.jpg') or file.endswith('.png')]

    num_examples = 5
    random_images = random.sample(image_files, num_examples)

    # display
    plt.figure(figsize=(15, 7))
    for i, image_file in enumerate(random_images):
        image_path = os.path.join(masked_folder, image_file)
        img = mpimg.imread(image_path)
        plt.subplot(1, num_examples, i+1)
        plt.imshow(img)
        plt.title(image_file)
        plt.axis('off')
    plt.show()

checkMaskedValidity(masked_folder=masked_face_folder)
```

```python
# Function to plot random images
def plot_random_images(resized_images, num_images=5):
    plt.figure(figsize=(15, 3))
    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        img = random.choice(resized_images)
        plt.imshow(img)
        plt.axis('off')
    plt.show()

# Print random images from with_mask_images
print("Random images from with_mask_images:")
plot_random_images(resized_images=with_mask_images)

# Print random images from without_mask_images
print("Random images from without_mask_images:")
plot_random_images(resized_images=without_mask_images)
```

```python
import random
import matplotlib.pyplot as plt

def checkLabelsMatchData(train_images, train_labels):
    # Choose a random index
    random_index = random.randint(0, len(train_images) - 1)

    # Retrieve the random image and its label
    random_image = train_images[random_index]
    random_label = train_labels[random_index]

    print(random_label)  # Print the label

    # Display the image
    plt.imshow(random_image)
    plt.axis('off')
    plt.show()

checkLabelsMatchData(train_images=train_images, train_labels=train_labels)
```
