```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_images, train_labels, test_size=0.2, random_state=42)
```

# Initial VGG19 Model Attempt

```python
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers

vgg_base = VGG19(weights='imagenet', include_top=False, input_shape=(224, 224, 3))


# the weights of layers will be updated. common practice when limited data
for layer in vgg_base.layers:
    layer.trainable = False
    
model = Sequential()
model.add(vgg_base)
# Add a GlobalAveragePooling2D layer instead of Flatten
model.add(GlobalAveragePooling2D())

# Dense layers for classification with batch normalization and dropout
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())

model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())

model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())

# Output layer for binary classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=12, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt

def printTrainingHistory():
    # Training History
    plt.figure(figsize=(8, 8))
    epochs_range = range(1, 21)  # For 20 Epochs
    plt.plot(epochs_range, history.history['accuracy'], label="Training Accuracy")
    plt.plot(epochs_range, history.history['val_accuracy'], label="Validation Accuracy")
    plt.axis(ymin=0.4, ymax=1)
    plt.grid()
    plt.title('Model Accuracy: VGG19')
    plt.ylabel('Accuracy')
    plt.xlabel('Epochs')
    plt.legend(['train', 'validation'])
    plt.show()
    
printTrainingHistory()
```

# Final Model Attempt
- Add L2 regualarizers to reduce overfitting of the data pipeline

```python
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers

vgg_base = VGG19(weights='imagenet', include_top=False, input_shape=(224, 224, 3))


# the weights of layers will be updated. common practice when limited data
for layer in vgg_base.layers:
    layer.trainable = False
    
model = Sequential()
model.add(vgg_base)
# Add a GlobalAveragePooling2D layer instead of Flatten
model.add(GlobalAveragePooling2D())

# Dense layers for classification with batch normalization and dropout
model.add(Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))  # Adding dropout for regularization

model.add(Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))  # Adding dropout for regularization

model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))  # Adding dropout for regularization

# Output layer for binary classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
# Print model summary
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=20, batch_size=20, validation_split=0.2)
```

```python
printTrainingHistory()
```

# Evaluative Statistics
- Tracemalloc to determine the memory consumption posed from the model

```python
import numpy as np
import time 
import tracemalloc

start_time = time.time()
tracemalloc.start()
# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)

end_time = time.time()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print('Current Memory Consumption: ', current / (1024**2), "Peak Memory Consumption: ", peak / (1024**2))
print('Time taken: ', end_time-start_time)
print('Test Accuracy =', accuracy)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Convert predicted probabilities to binary predictions (0 or 1)
y_pred_binary = (y_pred > 0.5).astype(int)
```

```python
import random
import matplotlib.pyplot as plt


def printTestCalculations(X_test, y_test, model):
    num_samples = 5

    # Randomly select indices of samples from X_test
    sample_indices = random.sample(range(len(X_test)), num_samples)
    
    fig, axes = plt.subplots(1, num_samples, figsize=(20, 4))  

    for i, index in enumerate(sample_indices):
        
        image = X_test[index]
        true_label = y_test[index]

        # predicted_probability being selected
        predicted_probability = model.predict(np.expand_dims(image, axis=0))[0]  

        predicted_probability_scalar = predicted_probability.item() if hasattr(predicted_probability, 'item') else predicted_probability
        predicted_label = 1 if predicted_probability_scalar > 0.5 else 0

        # display images
        axes[i].imshow(image)
        axes[i].axis('off')
        axes[i].set_title(f"True Label: {true_label}\nPredicted Label: {predicted_label} (Probability: {predicted_probability_scalar:.4f})", fontsize=10)

    plt.subplots_adjust(wspace=0.3)  
    plt.tight_layout()
    plt.show()
    
printTestCalculations(X_test=X_test, y_test=y_test, model=model)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def printConfusionMatrix(model, X_test, y_test):
    # Get predictions
    y_pred = model.predict(X_test)
    y_pred_classes = (y_pred > 0.5).astype(int)

    cm = confusion_matrix(y_test, y_pred_classes)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix: VGG19')
    plt.show()
printConfusionMatrix(model, X_test, y_test)
```

```python
from sklearn.metrics import classification_report, matthews_corrcoef

# Set the threshold for binary classification
threshold = 0.5

# Convert probabilities to binary predictions
y_pred_classes = (y_pred > threshold).astype(int)

# Generate classification report with handling for zero division
class_report = classification_report(y_test, y_pred_classes, zero_division=0)


mcc = matthews_corrcoef(y_test, y_pred_classes > 0.5)  
print("Matthews Correlation Coefficient (MCC):", mcc)

print("Classification Report:\n", class_report)
```

```python
def compute_complexity(model):
    complexity = 0
    for layer in model.layers:
        if isinstance(layer, tf.keras.layers.Dense):
            input_size = layer.input_shape[1]  # Number of input features
            output_size = layer.units  # Number of output units
            complexity += input_size * output_size
    return complexity

compute_complexity(model=model) # computation operations to process a single input 
```
