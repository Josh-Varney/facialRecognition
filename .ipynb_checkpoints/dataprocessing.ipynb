# Extra Preprocessing (Sample Selection and Normalisation of Data)

```python
import h5py

# Read from train_data.h5
with h5py.File('C:\\Users\\Jrv12\\Desktop\\Facial Recognition\\facialRecognition\\ifs\\train_data.h5', 'r') as hf:
    denoised_train_images = hf['denoised_train_images'][:]
    
# Read from train_labels.h5
with h5py.File('C:\\Users\\Jrv12\\Desktop\\Facial Recognition\\facialRecognition\\ifs\\train_labels.h5', 'r') as hf:
    train_labels = hf['train_labels'][:]
```

```python
indices_label_0 = np.where(train_labels == 0)[0]
sample_indices_label_0 = np.random.choice(indices_label_0, size=2500, replace=False)
```

- These images are then selected at random. The size of the dataset selected was _____

```python
import matplotlib.pyplot as plt

# Plot random images with label 0
num_samples_to_plot = 5  # You can adjust this number as needed
random_indices_label_0 = np.random.choice(indices_label_0, num_samples_to_plot, replace=False)

fig, axes = plt.subplots(1, num_samples_to_plot, figsize=(15, 3))  # Adjust figsize as needed
for i, idx in enumerate(random_indices_label_0):
    axes[i].imshow(denoised_train_images[idx], cmap='gray')  # Assuming grayscale images
    axes[i].set_title(f'Label: 0')
    axes[i].axis('off')  # Hide axis
plt.show()
```

- Investigating the widerspread of label 0 images without a masked individual

```python
indices_label_1 = np.where(train_labels == 1)[0]
sample_indices_label_1 = np.random.choice(indices_label_1, size=2500, replace=False)
```

- Selecting the images with a masked individual to select a sample for training and testing

```python
# Plot random images with label 1
num_samples_to_plot = 5  
random_indices_label_1 = np.random.choice(indices_label_1, num_samples_to_plot, replace=False)

fig, axes = plt.subplots(1, num_samples_to_plot, figsize=(15, 3))  # Adjust figsize as needed
for i, idx in enumerate(random_indices_label_1):
    axes[i].imshow(denoised_train_images[idx], cmap='gray')  # Assuming grayscale images
    axes[i].set_title(f'Label: 1')
    axes[i].axis('off')  # Hide axis
plt.show()
```

- Checking the validity of the sample before testing purposes

```python
# Use the sampled indices to select 1000 images and labels for each class
sample_train_label_0 = denoised_train_images[sample_indices_label_0]
sample_train_label_1 = denoised_train_images[sample_indices_label_1]
sample_labels_label_0 = train_labels[sample_indices_label_0]
sample_labels_label_1 = train_labels[sample_indices_label_1]

# Combine the samples from both classes
sample_train = np.concatenate([sample_train_label_0, sample_train_label_1], axis=0)
sample_labels = np.concatenate([sample_labels_label_0, sample_labels_label_1], axis=0)

# Randomize the order
random_indices = np.random.permutation(len(sample_train))
sample_train = sample_train[random_indices]
sample_labels = sample_labels[random_indices]
```

```python
sample_labels.shape
```

- Combination of the datasets and labels in preparation for training and testing.
- The training and testing sets `sample_train` and labels `sample_labels`

```python
import matplotlib.pyplot as plt

# Plot random samples
num_samples_to_plot = 5  # You can adjust this number as needed
random_indices = np.random.choice(len(sample_train), num_samples_to_plot, replace=False)

fig, axes = plt.subplots(1, num_samples_to_plot, figsize=(15, 3))  # Adjust figsize as needed
for i, idx in enumerate(random_indices):
    axes[i].imshow(sample_train[idx], cmap='gray')  # Assuming grayscale images
    axes[i].set_title(f'Label: {sample_labels[idx]}')
    axes[i].axis('off')  # Hide axis
plt.show()
```

- Checking the widespread of the sample test to see if there is even widespread of the classes for the binary classification model

```python
# Convert the image data type to float
sample_train = sample_train.astype(np.float32)

sample_train /= 255.0
```
